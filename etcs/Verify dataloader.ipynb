```python
from dataloader import Dataloader_COCO
from torch.utils.data import DataLoader
```

```python
cocoloader = Dataloader_COCO('test')
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
it = iter(cocoloader)
```

```python
img, bboxes = next(it)
fig, boxes = plt.subplots()
plt.imshow(img)
for bbox in bboxes:
    boxes.add_patch(Rectangle((bbox[0], bbox[1]), bbox[2], bbox[3], linewidth=1, edgecolor='r', facecolor='none'))
```
